```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math as mt
```

```python
def Gauss(A, b, xk):
    n = np.size(A)
    rx = b - np.dot(A,xk)
    print(rx)
    n_rx = np.linalg.norm(rx)
    print(n_rx)
    m = 1
    while m < 20:
        for i in range(0, 4):
            s = b[i]
                
            for j in range(0,4):
                #print(j)
                if j == i:
                    ind = j
                else:
                    s= s-(A[i,j]*xk[j])
            xi = s/A[i,i]
            xk[ind] = xi
            print(f'x{m},{i+1} = {xi}')
        rx = b - np.dot(A,xk)
        n_rx = np.linalg.norm(rx)
        print(f'x{m} = {xk} et sa norme = {n_rx}')
        m = m+1
```

```python
A = np.array([[6.0,-2.0,3.0,3.0],[16.0,-8.0,6.0,10.0],[3.0,-13.0,9.0,3.0],[-6.0,4.0,1.0,-18.0]])
b = np.array([16.0,26.0,-19.0,-34.0])
xk = np.array([2.0,2.0,2.0,2.0])
Gauss(A,b,xk)
```
